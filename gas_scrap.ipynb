```python

def get_info(path):
    driver.get(path)
    #make the page per100
    per100 = driver.find_element(By.ID, 'results_per_page')
    www = per100.find_elements(By.CSS_SELECTOR, 'option')
    but_100 = www[len(www)-1]
    but_100.click()
    time.sleep(3)
    
    get_holder = driver.find_element(By.ID, 'results')    
    #get the main able 
    main_table = get_holder.find_element(By.ID, 'list-results')
    heading = get_holder.find_elements(By.CSS_SELECTOR, 'thead td')
    heading_lst= []
    for i in heading:
        heading_lst.append(i.text)
    s = 0
#get the body li elements
    body_rows = main_table.find_elements(By.CSS_SELECTOR, 'tbody > tr')
    
    #lists for each column
    company = []
    reference = []
    city = []
    state = []
    description = []
    duration = []
    user_type = []
    min_val = []
    max_val = []
    w_2 = []
    updated =[]
    date_posted = []
    
    
    while True:
        s = s+1
        get_holder = driver.find_element(By.ID, 'results')
        main_table = get_holder.find_element(By.ID, 'list-results')
        body_rows = main_table.find_elements(By.CSS_SELECTOR, 'tbody > tr')
        for i,x in enumerate(body_rows):
            t_record = x.find_elements(By.CSS_SELECTOR, 'td')
            for ind, j in enumerate(t_record):
                if heading_lst[ind] == "Company":
                    company.append(j.text) 

                elif heading_lst[ind] == "City":
                    city.append(j.text)
                elif heading_lst[ind] == 'State':
                    state.append(j.text)
                elif heading_lst[ind] == 'Brief Description':
                    description.append(j.text)
                elif heading_lst[ind] == "Duration":
                    duration.append(j.text)
                elif heading_lst[ind] == 'User Type':
                    user_type.append(j.text)
                elif heading_lst[ind] == 'Min $':
                    min_val.append(j.text)
                elif heading_lst[ind] == 'Max $':
                    max_val.append(j.text)
                elif heading_lst[ind] == 'W-2':
                    w_2.append(j.text)
                elif heading_lst[ind] == 'Updated':
                    updated.append(j.text)
                elif heading_lst[ind] == 'Date Posted':
                    date_posted.append(j.text)

        try:
            get_ctrl = get_holder.find_element(By.CSS_SELECTOR, '.controls')
            #print('found it')
            next_buttons = get_ctrl.find_element(By.LINK_TEXT, "next")
            #print("u r doing great")
            if next_buttons.get_attribute("class"):
                break
            else:
                next_buttons.click()
                print("clicked")
            
        except:
            print("ooh something wrong")
            break
            
            
        #df_testing = pd.DataFrame(list(zip(company , city, state, description, duration, user_type, min_val, max_val, w_2,updated,date_posted)),columns=['Company','City', "State",'Brief Description','Duration','User Type','Min','Max','W-2','Updated', 'Date Posted'])
        #with pd.ExcelWriter("C:\\Users\\SHABA7\\DA-py-libs\\upwork\\CRNA\\GasWork_CRNA"+str(s)+".xlsx") as writer:
         #   df_testing.to_excel(writer, index=False)    
                
                
    zipped_lst = zip(company , city, state, description, duration, user_type, min_val, max_val, w_2,updated,date_posted)
        
    return zipped_lst
```
